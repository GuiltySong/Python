```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Baisic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

#resize the graph
plt.figure(figsize=(20,10), dpi = 200)

#plt.plot(x, y, label='2x', color = 'red', linewidth = '2', marker = '.', markeredgecolor = 'red')

#use shorthand notation
#fmt = '[color][marker][line]'

plt.plot(x, y, 'b^--', label='2x')


##LINE number 2
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:5], x2[:5]**2, 'r', label = 'x^2')
plt.plot(x2[4:], x2[4:]**2, 'r--')

plt.title('First Graph', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('X Axis', fontdict={'fontname':'Arial'})
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

plt.legend()

plt.savefig('mygraph.png', dpi=300)

plt.show()
```

### Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1, 4, 2]

bars = plt.bar(labels, values)

patterns = ['O', '/', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

###bars[0].set_hatch('O')
###bars[1].set_hatch('/')
###bars[2].set_hatch('*')

plt.figure(figsize=(6,4))
plt.show()
```
